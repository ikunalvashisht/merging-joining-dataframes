## Concatenation

Concatenation basically glues together DataFrames. Keep in mind that dimensions should match along the axis you are concatenating on. You can use **pd.concat** and pass in a list of DataFrames to concatenate together:

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']},
                        index=[10, 11, 12, 13])

df2 = pd.DataFrame({'P': ['A4', 'A5', 'A6','A7'],
                        'Q': ['B4', 'B5','B6', 'B7'],
                        'R': ['C4', 'C5', 'C6', 'C7'],
                        'S': ['D4', 'D5','D6', 'D7']},
                         index=[14, 15, 16, 17])

df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                        'B': ['B8', 'B9', 'B10', 'B11'],
                        'C': ['C8', 'C9', 'C10', 'C11'],
                        'D': ['D8', 'D9', 'D10', 'D11']},
                        index=[18, 19, 20, 21])
```

```python
df1
```

```python
frames = [df1,df2]
```

```python
pd.concat(frames) # here it did concatenation row-wise by default axis = 0 in the argument and join is outer
```

```python
pd.concat(frames, join = 'inner') # it did nothing when we change the join to inner 
# but it did change the index when we set ingore_index to True(it gave the pythonic index starting with 0)

#(boolean, default False. If True, do not use the index values on the concatenation axis. 
#(The resulting axis will be labeled 0, â€¦, n - 1.)
```

```python
result_concat = pd.concat(frames, keys = ['x','y','z'])
```

Pandas concat Vs append 

Concat gives the flexibility to join based on the axis( all rows or all columns)

Append is the specific case(axis=0, join='outer') of concat


When we concatenate DataFrames, we need to specify the axis. axis=0 tells Pandas to stack the second DataFrame under the first one. It will automatically detect whether the column names are the same and will stack accordingly. axis=1 will stack the columns in the second DataFrame to the RIGHT of the first DataFrame. To stack the data vertically, we need to make sure we have the same columns and associated column format in both datasets. When we stack horizonally, we want to make sure what we are doing makes sense (ie the data are related in some way)

```python
result_inner = pd.concat([df1,df4], axis = 0, join = 'inner', ignore_index = True)
```
